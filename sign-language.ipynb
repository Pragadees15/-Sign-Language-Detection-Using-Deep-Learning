```python

import kagglehub
import numpy as np
import pandas as pd
from PIL import Image
import os
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
import matplotlib.pyplot as plt
import seaborn as sns
import math
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix
import tensorflow as tf


# Define the root directory path
root_dir = "/kaggle/input/indian-sign-language-isl/Indian"

# Maximum number of images to load from each subdir_path
max_images_per_subdir = 500

# Initialize a list to store the image data
image_data = []

# Iterate through the subdirectories in the root directory
for label in os.listdir(root_dir):
    if label.isalnum():  # Include both letters and numbers
        subdir_path = os.path.join(root_dir, label)

        # Load a maximum of images and resize them
        for i, file in enumerate(os.listdir(subdir_path)):
            if i < max_images_per_subdir:
                img_path = os.path.join(subdir_path, file)
                try:
                    # Open, resize, and convert the image to RGB
                    img = Image.open(img_path).convert('RGB').resize((32, 32))  # Resize to (32, 32)
                    image_data.append({'label': label.strip(), 'color_image': img})  # Strip whitespace
                except Exception as e:
                    print(f"Error loading image {img_path}: {e}")

# Create a DataFrame from the image data
df = pd.DataFrame(image_data)

# Data Preprocessing
# Convert images to a consistent format
images = np.array([np.array(image['color_image']) for image in image_data])  # Extract and convert images
labels = np.array(df['label'])

# Normalize the pixel values
images = images / 255.0

# Label Encoding
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)  # Encode labels as integers

# Debugging: Print the encoded labels
print("Encoded labels:", encoded_labels)
print("Unique labels:", label_encoder.classes_)

# Ensure that the labels are of integer type
print("Labels data type:", encoded_labels.dtype)

# Splitting the data into Training and Testing
X_train, X_test, y_train, y_test = train_test_split(images, encoded_labels, test_size=0.2, random_state=42)

# Convert labels to categorical (one-hot encoding)
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Check shapes of the resulting datasets
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)

# Visualize label distribution
plt.figure(figsize=(10, 5))
sns.set_style("darkgrid")
sns.countplot(data=df, x='label')
plt.xlabel('Label')
plt.ylabel('Count')
plt.title('Distribution of Labels')
plt.show()

# Plot example images
num_labels = df['label'].nunique()
num_cols = 13
num_rows = math.ceil(num_labels / num_cols)

# Create subplots
fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 2))
axs = axs.flatten()  # Flatten the 2D array to 1D for easier indexing

# Plot the images
for i, (label, image) in enumerate(df.groupby('label').first().reset_index()[['label', 'color_image']].itertuples (index=False)):
    ax = axs[i]
    ax.imshow(image)
    ax.set_title('Label {}'.format(label))
    ax.axis('off')

# Hide any remaining empty subplots
for ax in axs[num_labels:]:
    ax.axis('off')

plt.tight_layout()
plt.show()

# Define the model architecture
model = Sequential([
    Conv2D(75, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Conv2D(50, (3, 3), activation='relu', padding='same'),
    Dropout(0.2),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Conv2D(25, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.3),
    Dense(len(np.unique(labels)), activation='softmax')  # Output layer with correct number of classes
])

model.compile(optimizer='SGD', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Data augmentation
datagen = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.1,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=False,
    vertical_flip=False
)

datagen.fit(X_train)

# Fitting the Model with stochastic process
learning_rate_reduction = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_accuracy', patience=2, verbose=1, factor=0.5, min_lr=0.001)

try:
    history = model.fit(datagen.flow(X_train, y_train), epochs=25, validation_data=(X_test, y_test), callbacks=[learning_rate_reduction])
except Exception as e:
    print(f"Error fitting the model: {e}")

# Model Evaluation
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
for i, metric in enumerate(['accuracy', 'loss']):
    axs[i].plot(history.history[metric], label='Training ' + metric.capitalize())
    axs[i].plot(history.history['val_' + metric], label='Testing ' + metric.capitalize(), linestyle='--')
    axs[i].set_xlabel('Epochs')
    axs[i].set_ylabel(metric.capitalize())
    axs[i].legend()
plt.tight_layout()
plt.show()

# Confusion Matrix
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

cm = confusion_matrix(y_true_classes, y_pred_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='d')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

accuracy = model.evaluate(X_test, y_test)
print(f'Validation test results - Loss: {accuracy[0]} - Accuracy: {accuracy[1]*100}%')

# Saving the Model
model.save("indianSignLanguage.h5")
```
